```python
merged_data = pd.concat([true, fake], ignore_index=True)
```

```python
shuffled_data = merged_data.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
shuffled_data.head(4)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.pipeline import make_pipeline
from sklearn.metrics import accuracy_score, classification_report
```

```python
train_data, test_data, train_labels, test_labels = train_test_split(
    shuffled_data['text'], shuffled_data['label'], test_size=0.2, random_state=42
)
```

```python
model = make_pipeline(CountVectorizer(), MultinomialNB())
```

```python
model.fit(train_data, train_labels)
```

```python
predictions = model.predict(test_data)
```

```python
accuracy = accuracy_score(test_labels, predictions)
report = classification_report(test_labels, predictions)
```

```python
print(f"Accuracy: {accuracy}")
print("Classification Report:\n", report)
```
