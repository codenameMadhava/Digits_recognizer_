```python
train_data = pd.read_csv("/kaggle/input/digit-recognizer/train.csv")
```

```python
train_data.head(5)
```

```python
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Conv2D,MaxPool2D,Flatten,Dense,Dropout,MaxPooling2D
import matplotlib.pyplot as plt
```

```python
test_data = pd.read_csv("/kaggle/input/digit-recognizer/test.csv")
test_data.head(5)
```

```python
x = train_data.drop('label',axis=1).values/255
y = train_data.label.values
test_data = test_data.values/255
```

```python
x = x.reshape(-1,28,28,1)
```

```python
test_data = test_data.reshape(-1,28,28,1)
```

```python
from keras.utils import to_categorical
```

```python
y = to_categorical(y,num_classes=10)
```

```python
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=.3,random_state=42)
```

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python

data_generator  = ImageDataGenerator(
    rotation_range=0,
    width_shift_range=0.0,
    height_shift_range=0.0,
    brightness_range=None,
    shear_range=0.0,
    horizontal_flip=False,
    vertical_flip=False,
)
```

```python
model = Sequential()
model.add(Conv2D(
    filters =28,
    kernel_size = (3,3),
    activation = 'relu',
    padding = 'valid',
    input_shape=(28,28,1),
    kernel_initializer = 'he_uniform'
))
model.add(MaxPooling2D(pool_size = (3,3)))
model.add(Flatten())
model.add(Dense(units=128,activation='relu'))
model.add(Dense(units=10,activation='softmax'))
```

```python
model.compile(
    loss= 'categorical_crossentropy',
    optimizer = 'adam',
    metrics = 'accuracy'
)
```

```python
history = model.fit(
    data_generator.flow(X_train,y_train,batch_size = 128),
    epochs = 20,
    validation_data = (X_test,y_test),
    steps_per_epoch = X_train.shape[0]/64
)
```

```python
y_predict = model.predict(X_test)
```

```python
predict = model.predict(test_data)
```

```python
predict_labels = np.argmax(predict,axis=1)
```

```python
pd.read_csv("/kaggle/input/digit-recognizer/sample_submission.csv")
```

```python
submit = pd.DataFrame({
    'ImageId':range(1,len(predict_labels)+1),
    'Label':predict_labels
})
```
